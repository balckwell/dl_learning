```python
import matplotlib.pyplot as plt
import random
from mplfonts import use_font
use_font('Noto Serif CJK SC')
plt.figure(figsize=(20, 8),dpi=200)

x = [1,2,3,5,8,13]
y = [4,7,11,22,5,7]
plt.plot(x, y)

plt.show()

```

三层结构：容器层: Canvas(画板，放置figure的工具),figures（画纸）,Axes
Axes:figures:通过plt.figure（）设置画布的大小和分辨率等等
     Axes(坐标系)：数据的绘图区域
     Axis(坐标轴)：坐标系中的一条轴，包含大小限制、刻度和刻度标签

辅助显示层：包括Axes外观（facecolor），边框线（spines）,坐标轴名称（axis label）,坐标轴刻度（tick）,坐标轴刻度标签（tick label）,网格线（grid）,图例（legend）,标题（title）

图像层：
指Axes通过plot,scatter,bar, histogram, pie等函数绘制的图像

#3案例：显示温度变化情况

```python
#0.生成数据

x = range(60)
y_beijing =[random.uniform(10, 15) for i in x]
y_shanghai = [random.uniform(15, 25) for i in x]

#1.创建画布
plt.figure(figsize = (20, 8), dpi=100)

#2.图形绘制
plt.plot(x,y_beijing,label="北京",color="r", linestyle="-.")
plt.plot(x, y_shanghai,label="上海")
#2.1添加x,y刻度
y_ticks  =range(40)

x_ticks_labels = ["11时{}分".format(i) for i in x]
plt.xticks(x[::5], x_ticks_labels[::5])
#必须最开始传进去的是数字

#2.2添加网格
plt.grid(True, linestyle="--", alpha=1)
plt.yticks(y_ticks[0:40:5])#[::5]

#2.3添加描述
plt.xlabel("时间",fontsize=10)
plt.ylabel("温度",fontsize=10)
plt.title("一小时温度变化图",fontsize=20)

#2.4显示图例
plt.legend(loc = 0)
#3.图像显示
plt.show()

```
